### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python

loja['Faturamento'] = loja['Preço'] * loja['Quantidade de parcelas']
loja2['Faturamento'] = loja2['Preço'] * loja2['Quantidade de parcelas']
loja3['Faturamento'] = loja3['Preço'] * loja3['Quantidade de parcelas']
loja4['Faturamento'] = loja4['Preço'] * loja4['Quantidade de parcelas']

faturamento_lojas = {
    "Loja 1": loja['Faturamento'].sum(),
    "Loja 2": loja2['Faturamento'].sum(),
    "Loja 3": loja3['Faturamento'].sum(),
    "Loja 4": loja4['Faturamento'].sum()
}

faturamento_lojas
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(faturamento_lojas.keys(), faturamento_lojas.values())
plt.title("Faturamento por Loja")
plt.xlabel("Lojas")
plt.ylabel("Faturamento")
plt.show()
```

# 2. Vendas por Categoria

```python
categoria_loja1 = loja['Categoria do Produto'].value_counts()
categoria_loja2 = loja2['Categoria do Produto'].value_counts()
categoria_loja3 = loja3['Categoria do Produto'].value_counts()
categoria_loja4 = loja4['Categoria do Produto'].value_counts()

categoria_loja1
```

```python
categoria_loja1.plot(kind='bar', title='Vendas por Categoria - Loja 1')
plt.xlabel("Categoria")
plt.ylabel("Quantidade")
plt.show()
```

# 3. Média de Avaliação das Lojas


```python
avaliacoes = {
    "Loja 1": loja['Avaliação da compra'].mean(),
    "Loja 2": loja2['Avaliação da compra'].mean(),
    "Loja 3": loja3['Avaliação da compra'].mean(),
    "Loja 4": loja4['Avaliação da compra'].mean()
}

avaliacoes
```

```python

plt.figure(figsize=(8,5))
plt.bar(avaliacoes.keys(), avaliacoes.values())
plt.title("Avaliação Média por Loja")
plt.xlabel("Lojas")
plt.ylabel("Avaliação")
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
mais_vendido_loja1 = loja['Produto'].value_counts().idxmax()
menos_vendido_loja1 = loja['Produto'].value_counts().idxmin()

print("Mais vendido:", mais_vendido_loja1)
print("Menos vendido:", menos_vendido_loja1)
```

# 5. Frete Médio por Loja

```python
frete_medio = {
    "Loja 1": loja['Frete'].mean(),
    "Loja 2": loja2['Frete'].mean(),
    "Loja 3": loja3['Frete'].mean(),
    "Loja 4": loja4['Frete'].mean()
}

frete_medio
```

```python
plt.figure(figsize=(8,5))
plt.bar(frete_medio.keys(), frete_medio.values())
plt.title("Frete Médio por Loja")
plt.xlabel("Lojas")
plt.ylabel("Frete Médio")
plt.show()
```

# **Relatório Final**

**Introdução **

O objetivo desta análise foi avaliar o desempenho das quatro lojas da rede Alura Store para auxiliar o Senhor João na decisão de qual unidade vender para iniciar um novo empreendimento. Para isso, foram analisados diferentes indicadores de desempenho, incluindo faturamento total, categorias de produtos mais vendidas, avaliação média dos clientes, produtos com maior e menor volume de vendas e o valor médio do frete. A análise foi realizada a partir de dados de vendas e visualizações gráficas, permitindo uma comparação clara entre as lojas.

**Desenvolvimento da Análise**

A primeira métrica analisada foi o faturamento total de cada loja, obtido a partir da soma das vendas realizadas. Essa análise permite identificar quais lojas possuem maior capacidade de geração de receita. Ao comparar os valores de faturamento entre as quatro lojas, foi possível observar diferenças relevantes no desempenho financeiro.

Em seguida, foram analisadas as categorias de produtos mais vendidas, permitindo entender quais tipos de produtos possuem maior demanda em cada loja. Essa informação é importante porque revela o perfil de consumo dos clientes e o desempenho das categorias dentro de cada unidade.

Outro indicador relevante foi a média das avaliações dos clientes. As avaliações refletem diretamente a satisfação dos consumidores com a experiência de compra, incluindo fatores como qualidade do produto, atendimento e entrega. Lojas com médias mais altas tendem a apresentar maior fidelização de clientes e melhor reputação.

Também foram identificados os produtos mais e menos vendidos, o que ajuda a compreender quais itens possuem maior aceitação no mercado e quais apresentam baixa demanda. Esse tipo de análise pode indicar oportunidades de melhoria no mix de produtos ou na estratégia de vendas.

Por fim, foi analisado o frete médio das lojas, um fator que pode impactar diretamente a decisão de compra dos clientes. Custos de frete mais elevados podem reduzir a competitividade da loja em comparação com outras unidades ou concorrentes.

**Conclusão e Recomendação**

Com base nas análises realizadas, é possível identificar que a Loja 4 apresenta desempenho inferior em comparação às demais unidades em alguns dos indicadores avaliados, como faturamento total, avaliações dos clientes ou custo médio de frete.

Considerando esses fatores, recomenda-se que o Senhor João considere a venda dessa loja para direcionar seus investimentos em um novo empreendimento. Essa decisão permite manter as lojas com melhor desempenho operacional e potencial de crescimento, garantindo maior eficiência na gestão da rede.
